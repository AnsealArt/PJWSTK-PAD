### Import libraries

```python
import pandas as pd
```

### Read orders

```python
df_orders = pd.read_csv("orders.csv")
df_orders
```

### Convert order_date

```python
df_orders['order_date'] = pd.to_datetime(df_orders['order_date'], format="%Y-%m-%d")
df_orders['order_date'].sample(1)
```

### T-Shirt Category analysis

```python
df_orders['tshirt_category'].value_counts()
```

### T-Shirt Category transformation

```python
# df_orders['tshirt_category'] = df_orders['tshirt_category'].replace({
#     "^Tshirt$" : "T-Shirt",
#     "^Shirt$" : "T-Shirt",
#     "^Wh$" : "White",
#     "^Bl$" : "Black"
# }, regex=True)

df_orders['tshirt_category'] = df_orders['tshirt_category'].apply(
    lambda item : item\
        .replace("Tshirt", "T-Shirt")\
        .replace("Wh T-Shirt M", "White T-Shirt M")\
        .replace("Bl T-Shirt M", "Black T-Shirt M")\
        .replace("Wh T-Shirt F", "White T-Shirt F")\
        .replace("Bl T-Shirt F", "Black T-Shirt F")\
        .replace("Tennis Shirt", "Unknown Tennis T-Shirt")\
        .replace("Hoodie", "Unknown Hoodie Unknown")
)

df_orders['tshirt_category'].value_counts()
```

### T-Shirt data extraction

```python
df_orders['tshirt_colour'] = df_orders['tshirt_category'].apply(lambda item : item.split(" ", 3)[0])
df_orders['tshirt_type'] = df_orders['tshirt_category'].apply(lambda item : item.split(" ", 3)[1])
df_orders['tshirt_size'] = df_orders['tshirt_category'].apply(lambda item : item.split(" ", 3)[2])
df_orders
```

### Find orders created between 31-12-2014 and 02-08-2016.

```python
df_orders[(df_orders['order_date'] >= '2014-12-31') & (df_orders['order_date'] <= '2016-08-02')].sort_values(by='order_date')
```
